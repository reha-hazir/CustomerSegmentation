```python
df = pd.read_csv('czech-financial-dataset/data/trans.csv')
df.head()
```

```python
#drop columns
drop_cols = [
    'k_symbol',
    'bank',
    'account'
]
df = df.drop(drop_cols, axis=1)

#remove missing values
df.dropna(subset=['operation'], inplace=True)

#rename columns
df = df.rename(columns={'type': 'TransType', 'operation': 'TransOp','date':'TransDate'})

#Change the czech terms into english
mapping = {'PRIJEM':'INCOME', 'VYDAJ':'EXPENSE', 'VYBER':'WITHDRAWAL'}
mapping2 = {'VKLAD': 'depositing money', 'PREVOD Z UCTU':'transfer from account', 'VYBER':'withdrawing money', 
            'PREVOD NA UCET':'transfer to account','VYBER KARTOU':'withdrawing by card'}
df['TransType'] = df['TransType'].replace(mapping)
df['TransOp'] = df['TransOp'].replace(mapping2)

#Changing the date column into DateTime Object
df['TransDate'] = pd.to_datetime(df['TransDate'], format='%y%m%d')
```

```python
df.head()
```

```python
df['TransType'].unique()
```

```python
data = []
for account in df['account_id'].unique():
    x = df[df['account_id'] == account]
    income = x[x['TransType'] == 'INCOME']['amount'].sum()
    expense = x[x['TransType'] == 'EXPENSE']['amount'].sum()
    withdraw = x[x['TransType'] == 'WITHDRAWAL']['amount'].sum()
    d = {'AccId': account, 'IncomeSum': income , 'ExpenseSum': expense, 'WithdrawSum': withdraw}
    data.append(d)

df = pd.DataFrame(data=data)
df.head()
```
